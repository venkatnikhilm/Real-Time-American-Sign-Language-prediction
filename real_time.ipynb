```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
```

```python
# Load the trained model
model = load_model('sign_mnist_model.h5')

# Labels mapping (adjust based on your dataset)
labels = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E', 5: 'F', 6: 'G', 7: 'H', 8: 'I',
          9: 'K', 10: 'L', 11: 'M', 12: 'N', 13: 'O', 14: 'P', 15: 'Q', 16: 'R',
          17: 'S', 18: 'T', 19: 'U', 20: 'V', 21: 'W', 22: 'X', 23: 'Y'}
```

```python
def preprocess_frame(frame):
    """Convert frame to grayscale, resize, normalize, and reshape."""
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    resized = cv2.resize(gray, (28, 28))  # Resize to 28x28
    normalized = resized / 255.0  # Normalize pixel values
    reshaped = normalized.reshape(1, 28, 28, 1)  # Reshape for model input
    return reshaped
```

```python
# Open the webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

print("Webcam initialized. Press 'q' to quit.")
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Flip frame for a mirrored view
    frame = cv2.flip(frame, 1)

    # Define a larger region of interest (ROI)
    cv2.rectangle(frame, (150, 150), (500, 500), (255, 0, 0), 2)  # Adjusted coordinates
    roi = frame[150:500, 150:500]  # Larger box coordinates

    # Preprocess the ROI for the model
    processed = preprocess_frame(roi)

    # Make a prediction
    prediction = model.predict(processed)
    predicted_label = np.argmax(prediction)
    predicted_char = labels.get(predicted_label, "?")  # Fallback to "?" if not found

    # Display the prediction on the video feed
    cv2.putText(frame, f"Prediction: {predicted_char}", (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Show the video feed
    cv2.imshow('Sign Language Recognition', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
cap.release()
cv2.destroyAllWindows()
```
